```python
# Importing necessary libraries 

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('bmh')
from matplotlib.pyplot import figure

%matplotlib inline

```

```python
# Loading the csv files into the notebook

price_data_df = pd.read_csv(r"C:\Users\rauna\Downloads\price_data (1).csv")
price_data_df.head()

client_data_df = pd.read_csv(r"C:\Users\rauna\Downloads\client_data (1).csv")
client_data_df.head()
```

```python
# Generating Descriptive Statistics

price_data_df.dtypes
price_data_df.describe()
price_data_df.info()

client_data_df.dtypes
client_data_df.describe()
client_data_df.info()
```

```python
# Visualising the data

sns.set(style = "darkgrid")

# For numerical columns using Box plots and Histograms.



#Histograms for numerical columns in price_data and client data

price_num_columns = ['price_off_peak_var', 'price_peak_var', 'price_mid_peak_var', 'price_off_peak_fix', 'price_peak_fix', 'price_mid_peak_fix']
price_data_df[price_num_columns].hist(bins = 30)
plt.suptitle('Histograms of Numerical Columns in price_data')
plt.show()

client_num_columns = ['cons_12m', 'cons_gas_12m', 'cons_last_month', 'forecast_cons_12m', 'forecast_cons_year', 'forecast_discount_energy', 'forecast_meter_rent_12m', 'forecast_price_energy_off_peak', 'forecast_price_energy_peak', 'forecast_price_pow_off_peak', 'imp_cons', 'margin_gross_pow_ele', 'margin_net_pow_ele', 'nb_prod_act', 'net_margin', 'num_years_antig', 'pow_max']
client_data_df[client_num_columns].hist(bins=30 , figsize = (20,15))
plt.suptitle('Histograms of Numerical Columns in client_data')
plt.show()

```

```python
# Box plots for numerical columns in price_data

plt.figure(figsize=(12, 8))
sns.boxplot(data=price_data_df[price_num_columns])
plt.xticks(rotation=90)
plt.title('Box Plots of Numerical Columns in price_data')
plt.show()

# Box plots for numerical columns in client_data
plt.figure(figsize=(12,8))
sns.boxplot(data=client_data_df[client_num_columns])
plt.xticks(rotation=90)
plt.title('Box Plots of Numerical Columns in client_data')
plt.show()


```

```python
# Churn status
churn = client_data_df[['id', 'churn']]
churn.columns = ['Companies', 'churn']
churn_total = churn.groupby(churn['churn']).count()
churn_percentage = churn_total / churn_total.sum() * 100

# Plot Stacked Bar for Churn Status
churn_percentage.transpose().plot(kind="bar", stacked=True, figsize=(5, 5))
plt.title("Churning status")
plt.ylabel("Company base (%)")
plt.legend(["Retention", "Churn"], loc="lower right")
for p in plt.gca().patches:
    plt.gca().annotate(str(round(p.get_height(), 1)), (p.get_x() * 1.005, p.get_height() * 1.005))
plt.show()
```

```python


# Sales channel and churn status
channel = client_data_df[['id', 'channel_sales', 'churn']]
channel = channel.groupby([channel['channel_sales'], channel['churn']])['id'].count().unstack(level=1).fillna(0)
channel_churn = (channel.div(channel.sum(axis=1), axis=0) * 100).sort_values(by=[1], ascending=False)

# Plot Stacked Bar for Sales Channel
channel_churn.plot(kind="bar", stacked=True, figsize=(18, 10), rot=30)
plt.title('Sales channel')
plt.ylabel("Company base (%)")
plt.legend(["Retention", "Churn"], loc="upper right")
for p in plt.gca().patches:
    plt.gca().annotate(str(round(p.get_height(), 1)), (p.get_x() * 1.005, p.get_height() * 1.005))
plt.show()
```
